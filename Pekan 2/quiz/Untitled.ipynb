```python
data11 = {
    'food' : ['seblak', 'cireng', 'cimol', 'somay', 'batagor'],
    'price' : [5, 5, 8, 8, 6],
    'quantity' : [2, 2, 2, 2, 4],
}

data12 = {
    'food' : ['seblak', 'cireng', 'dimsum', 'somay', 'batagor'],
    'netto' : [4, 2, 3, 3, 2],
}

df11 = pd.DataFrame(data11)
df12 = pd.DataFrame(data12)
```

```python
df11
```

```python
df12
```

```python
df4=pd.merge(df11,df12,on='food',how='inner')
df4
```

```python
df4=pd.merge(df11,df12,on='food',how='left')
df4
```

```python
df4=pd.merge(df11,df12,on='food',how='right')
df4
```
