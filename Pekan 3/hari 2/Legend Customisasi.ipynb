```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
cos = np.cos(x)
sin = np.sin(x)
sin2 = np.sin(x*2)
```

```python
plt.style.use('seaborn')

fig, ax = plt.subplots()

ax.plot(x, cos)
ax.plot(x, sin)

plt.show()
```

```python
# digunakan custom legend
plt.style.use('seaborn')

fig, ax = plt.subplots()

ax.plot(x, cos, linestyle='--')
ax.plot(x, sin)

ax.legend()

plt.show()
```

```python
# digunakan custom legend
plt.style.use('seaborn')

fig, ax = plt.subplots()

ax.plot(x, cos, linestyle='--', label='CosX')
ax.plot(x, sin, label='SinX')

ax.legend()

plt.show()
```

```python
# digunakan custom legend
plt.style.use('seaborn')

fig, ax = plt.subplots()

ax.plot(x, cos, linestyle='--', label='CosX')
ax.plot(x, sin, label='SinX')
ax.plot(x, sin2, label='Sin2X')

ax.legend(loc='lower left', title='Grafik Cos Sin')

plt.show()
```
