```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# scatter plot
iris = sns.load_dataset('iris')
```

```python
# seaborn bisa langsung memasukkan data-nya dengan dataframe
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')
plt.show()
```

```python
# data kategori
sns.countplot(data=iris, x='species')
```

```python
# membaginya berdasarkan warna
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species') # hue harus kategori
plt.show()
```

```python
flight = sns.load_dataset('flights')
```

```python
flight.head()
```

```python
flight.year.unique()
```

```python
# line plot
sns.relplot(data=flight, x='year', y='passengers', kind='line', alpha=0.4) # semacam di groupby
plt.show()
```

```python
# line plot (ada central tendency, ada distribusi data-nya, dan lebih penting kita lihat trend data-nya)
sns.relplot(data=flight, x='year', y='passengers', kind='line', alpha=0.4, ci='sd') # line --> rata-rata, shade --> standar deviasi
plt.show()
```

```python
# line plot
sns.relplot(data=flight, x='year', y='passengers', kind='line', hue='month') # trend dalam tahun, yang dibagi-bagi dalam bulan
plt.show()
```

```python
# scatter plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', alpha=0.4, style='day')
plt.show()
```

```python
# scatter plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', alpha=0.4, style='day', hue='time')
plt.show()
```

```python
tips.head()
```

```python
# cat plot
sns.catplot(x='time', data=tips, kind='count')
plt.show()
```

```python
# cat plot
sns.catplot(x='time', data=tips, kind='count', col='smoker') # menambahkan subplot
plt.show()
```

```python
# bar plot (beda count plot dan bar plot)
sns.catplot(x='time', y='total_bill', data=tips, kind='bar')
plt.show()
```

```python
# bar plot --> summary, terdapat central tendency
sns.catplot(x='time', y='total_bill', data=tips, kind='bar', ci='sd') # ditambahkan standar deviasi
plt.show()
```

```python
# box plot
sns.catplot(x='time', y='total_bill', data=tips, kind='box')
plt.show()
```
