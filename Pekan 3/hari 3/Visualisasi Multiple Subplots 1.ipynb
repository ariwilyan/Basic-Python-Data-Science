```python
# menempatkan multiple data sebelumnya ke multiple axis, dengan tiap data memiliki 1 axis
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

ax[0][0].plot(x, cos, color='b', label='cosx')
ax[0][1].plot(x, cos3, color='g', label='3cosx')
ax[1][0].plot(x, sin, color='y', label='sin90')
ax[1][1].plot(x, sin2, color='r', label='2sinx')

fig.legend()

plt.show()
```

**Note** : Apabila dilihat, mungkin terlihat sama untuk hasilnya. Namun apabila lebih diteliti lagi, terdapat perbedaan untuk rentang nilai pada sumbu y pada beberapa plot. Sehingga disini proses analisis yang mungkin nantinya akan dilakukan tidak sesuai untuk semua plot-nya. Maka perlu disamakan terlebih dahulu agar semua plot memiliki rentang nilai pada sumbu x dan sumbu y dengan nilai yang sama.

```python
# memanfaatkan parameter lainnya, yaitu sharex ataupun sharey menyesuiakan dengan rentang mana yang ingin disamakan antar plot
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharey=True)

ax[0][0].plot(x, cos, color='b', label='cosx')
ax[0][1].plot(x, cos3, color='g', label='3cosx')
ax[1][0].plot(x, sin, color='y', label='sin90')
ax[1][1].plot(x, sin2, color='r', label='2sinx')

fig.legend()

plt.show()
```

```python
# dapat pula disesuaikan untuk axis x nya, atau sumbu x nya dengan menambahkan parameter sharex
# namun, karena pada sumbu x atau axis x nya sudah memiliki rentang nilai yang sama, tidak perlu dilakukan set parameter sharex nya
# karena useless, tidak akan berpengaruh apa-apa. Hasilnya akan tetap sama!
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharey=True, sharex=True)

ax[0][0].plot(x, cos, color='b', label='cosx')
ax[0][1].plot(x, cos3, color='g', label='3cosx')
ax[1][0].plot(x, sin, color='y', label='sin90')
ax[1][1].plot(x, sin2, color='r', label='2sinx')

fig.legend()

plt.show()
```
