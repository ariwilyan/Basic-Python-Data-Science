```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# membuat axes di dalam axes
fig, ax = plt.subplots(figsize=(16, 8), nrows=1, ncols=2)
```

```python
# membuat axes di dalam axes dengan menggunakan method .inset_axes(), method ini dimiliki oleh objek axes

fig, ax = plt.subplots(figsize=(16, 8), nrows=1, ncols=2)
ax[0].inset_axes([0.8, 0.8, 0.3, 0.3])
```

```python
# membuat axes di dalam axes dengan menggunakan method .inset_axes(), method ini dimiliki oleh objek axes

fig, ax = plt.subplots(figsize=(16, 8), nrows=1, ncols=2)

ax[0].plot(x, cos, color='b')
ins1 = ax[0].inset_axes([0.8, 0.8, 0.3, 0.3])
ins1.plot(x, sin, color='y')

ax[0].set_title('Grafik Cosx')
ins1.set_title('Grafik Sinx')

ax[1].plot(x, cos3, color='g')
ins2 = ax[1].inset_axes([-0.2, -0.2, 0.3, 0.3])
ins2.plot(x, sin2, color='r')

plt.show()
```
