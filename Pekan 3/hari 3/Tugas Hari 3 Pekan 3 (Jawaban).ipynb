```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16, 8), sharey=True)

ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, sin, color='r', label='sinX')
ax[2].plot(x, cos, color='g', label='cosX')

ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(15, 8), sharex=True, sharey=True)

ax.plot(x, y)

# multiplot in upper right side
ins_upper_right = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins_upper_right.plot(x, y, color='blue')
ins_upper_right.plot(x, y4, color='black')

# multiplot in upper left side
ins_upper_left = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins_upper_left.plot(x, y, color='blue')
ins_upper_left.plot(x, y5, color='y')

# multiplot in lower right side
ins_lower_right = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins_lower_right.plot(x, y, color='blue')
ins_lower_right.plot(x, y3, color='g')

# multiplot in lower left side
ins_lower_left = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins_lower_left.plot(x, y, color='blue')
ins_lower_left.plot(x, y2, color='r')

ax.set_title('Grafik CosX')
ins_upper_right.set_title('Grafik CosX^4')
ins_upper_left.set_title('Grafik CosX^5')
ins_lower_right.set_title('Grafik CosX^3')
ins_lower_left.set_title('Grafik CosX^2')

plt.show()
```
