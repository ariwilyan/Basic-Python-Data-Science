# Scatter

**Scatter** : salah satu bentuk visualisasi data, dimana kita memvisualisasikan dua varibel bertipe data **_numerik_**. Ketika kita memvisualisasikan data dengan scatter plot, hasilnya akan berbentuk **data point**.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y) # secara default, ax.plot akan berguna untuk membuat data berupa line plot. Untuk menjadikannya scatter plot, perlu ditambahkan beberapa argumen lagi. Seperti pada contoh code selanjutnya.

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(x, y, marker='v')

plt.show()
```

```python
# tambahkan 1 argumen lagi untuk menjadikannya scatter plot
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(x, y, marker='v', linestyle='None')

plt.show()
```

```python
# tambahkan 1 argumen lagi untuk menjadikannya scatter plot
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(x, y, linestyle='None')

plt.show()
```

## Menggunakan Method Scatter

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(x, y, marker='v')

plt.show()
```

## Perbedaan Method Scatter dengan Plot

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(x, y, marker='v', s=10) # s: untuk mengubah ukuran dari tiap data point

plt.show()
```

```python
# kalau plot tidak memiliki argumen tersebut
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(x, y, marker='v', linestyle='None', s=100)

plt.show()
```

## Customisasi Scatter Plot

```python
rnd = np.random.RandomState(30)
x = rnd.randn(100)
y = rnd.randn(100)
colors = rnd.rand(100)
sizes = 10000*rnd.rand(100)
```

```python
# kita lihat hasil customisasi scatter plot tersebut
fig, ax = plt.subplots(figsize=(12, 7))

# ax.scatter(x, y, c=colors, s=sizes, alpha=0.5) # alpha memberikan transparansi terhadap data point
# ax.scatter(x, y, c=colors, s=sizes, alpha=1)
mappbar = ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='summer') # cmap bisa dilihat pada dokumentasi matplotlib
fig.colorbar(mappbar)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

# membuat scatter plot dengan filter reading score untuk male, dan writing score untuk female. Ini difilter berdasarkan column gender
rs_male = df[df['gender'] == 'male']['reading score']
ws_male = df[df['gender'] == 'male']['writing score']

rs_female = df[df['gender'] == 'female']['reading score']
ws_female = df[df['gender'] == 'female']['writing score']

ax.scatter(rs_male, ws_male, color='b', label='Male')
ax.scatter(rs_female, ws_female, color='y', label='Female')

ax.set_xlabel('Reading Score')
ax.set_ylabel('Writing Score')
ax.set_title('Student Performance')

plt.legend()

plt.show()
```

```python
# data dengan gender male saja
df[df['gender'] == 'male']
```

```python
# data reading score untuk gender male saja
df[df['gender'] == 'male']['reading score']
```

```python
# plot math score, reading score, dan writing score dalam 1 axes
fig, ax = plt.subplots(figsize=(12, 8))

MS = df['math score']
RS = df['reading score']
WS = df['writing score']

axmap = ax.scatter(RS, WS, c=MS, cmap='PuBuGn') # cmap : ganti tema warna data point

ax.set_xlabel('Reading Score')
ax.set_ylabel('Writing Score')
ax.set_title('Student Performance')

fig.colorbar(axmap)

plt.show()
```
