## Cleaning Data

```python
cols = ['DewPointAvgF', 'HumidityAvgPercent', 'TempAvgF', 'WindAvgMPH']
df = df[cols]
```

```python
df.head()
```

```python
df.info()
```

```python
df['DewPointAvgF'].astype('float')
```

```python
# mengubah '-' menjadi Nan
df = df.replace('-', np.nan)
```

```python
# menjumlahkan data null/NaN dari setiap column data kita
df.isnull().sum()
```

```python
df.fillna?
```

```python
# mengisi nan value dengan ffill
df = df.fillna(method='ffill')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['DewPointAvgF'] = df['DewPointAvgF'].astype('float')
df['HumidityAvgPercent'] = df['HumidityAvgPercent'].astype('float')
df['WindAvgMPH'] = df['WindAvgMPH'].astype('float')
```

```python
df.columns
```

**Note** :

Di matplotlib, data yg dimasukkan untuk diplot bisa berupa series dan dataframe

```python
# memvisualisasi
plt.figure(figsize=(15, 10))

plt.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], c=df['TempAvgF'], cmap='coolwarm', s=df['WindAvgMPH']*20, alpha=0.4)

plt.title('Austin Weather')
plt.xlabel('Humadity Avg %')
plt.ylabel('Dew Point Avg (F)')
plt.colorbar()

plt.show()
```

<h2>Soal 2: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
# Grouping Data
df_group = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
```

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

```python
x = np.array([1, 4, 7, 10])
# dibuat untuk tiap group bar ukurannya 2.4 --> dengan tiap bar ukurannya 0.2 (satu groupbar fullnya ukuran sebesar 3.0)
```

```python
# versi kalau memasukkan datanya satu-satu
#code here
fig, ax = plt.subplots(figsize=(16, 8))
ax.bar(x, df_group.iloc[0, :], label=df_group.iloc[0, :].name, width=0.2)
ax.bar(x+0.2, df_group.iloc[1, :], label=df_group.iloc[1, :].name, width=0.2)
ax.bar(x+0.4, df_group.iloc[2, :], label=df_group.iloc[2, :].name, width=0.2)

plt.legend()

plt.show()
```

```python
# versi looping dengan iterrows
#code here
plt.style.use('seaborn')
x_offset = 0.2
fig, ax = plt.subplots(figsize=(16, 8))

for row in df_group.iterrows():
    ax.bar(x+x_offset, row[1], label=row[0], width=0.2)
    x_offset += 0.2

ax.set_title('Mean Sales Video Games By Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_xticks(x+1.2)
ax.set_xticklabels(df_group.columns)

plt.legend()

plt.show()
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
df_group.iloc[0, :]
```

```python
# versi no-loop (namun akan lama, maka akan lebih baik menggunakan for-loop)
#code here
plt.style.use('ggplot')
x = [1, 2, 3, 4]
bottom = [0, 0, 0, 0]

fig, ax = plt.subplots(figsize=(16, 8))

ax.bar(x, df_group.iloc[0, :], label=df_group.iloc[0, :].name, bottom=bottom)
ax.bar(x, df_group.iloc[1, :], label=df_group.iloc[1, :].name, bottom=bottom+df_group.iloc[0, :])
ax.bar(x, df_group.iloc[2, :], label=df_group.iloc[2, :].name, bottom=bottom+df_group.iloc[0, :]+df_group.iloc[1, :])
ax.bar(x, df_group.iloc[3, :], label=df_group.iloc[3, :].name, bottom=bottom+df_group.iloc[0, :]+df_group.iloc[1, :]+df_group.iloc[2, :])

ax.set_title('Mean Sales Video Games By Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_xticks(x)
ax.set_xticklabels(df_group.columns)

plt.legend()

plt.show()
```

```python
# versi looping dengan iterrows
#code here
plt.style.use('ggplot')
x = [1, 2, 3, 4]
bottom = [0, 0, 0, 0]

fig, ax = plt.subplots(figsize=(16, 8))

for row in df_group.iterrows():
    ax.bar(x, row[1], label=row[0], bottom=bottom)
    bottom += row[1]

ax.set_title('Mean Sales Video Games By Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
ax.set_xticks(x)
ax.set_xticklabels(df_group.columns)

plt.legend()

plt.show()
```
