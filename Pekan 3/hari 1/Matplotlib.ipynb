### Style

```python
# cara mengganti atau menggunakan style di matplotlib
plt.style.use('classic')
```

```python
# cara melihat style yang available di matplotlib
plt.style.available
```

```python
# secara default style di matplotlib seperti ini,
import numpy as np

x = np.linspace(0, 10, 100) # membuat array dari 0-10, sebanyak 100 point
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
# tanpa menggunakan plt.show() akan tetap muncul gambar visualisasi plotnya
import numpy as np

x = np.linspace(0, 10, 100) # membuat array dari 0-10, sebanyak 100 point
y = np.cos(x)
plt.plot(x, y)
# plt.show() #
```

**Note**:

tanpa show sebenarnya tetap akan ditampilkan plotnya, namun sebenarnya show ini adalah event loop, jadi dia akan melihat di kode kita apakah ada objek figure atau tidak. Kalau ada objek figure yang aktif, maka dia akan memanggil figure itu kemudian menampilkannya.

jadi berguna atau tidaknya plt.show() ini bergantung dari konteks python kita dijalankan dalam media apa. Ada 3 media dalam menjalankan python:
1. Bisa di dalam script --> script disini seperti pada biasanya, yang ada pada python comand prompt atau langsung lewat teks editor
2. Bisa di ipython --> singkatan dari **Interactive Python**, bisa dicek lewat cmd, kemudian ketikkan ipython. Di ipython ini, diharuskan menggunakan plt.show() untuk bisa ditampilkan visualisasi plotnya. Namun bisa juga digantikan dengan menggunakan tambahan script **%matplotlib**. Namun kekurangannya, apabila menggunakan %matplotlib, setelah kita melakukan plt.plot atau dimasukkannya data untuk divisualisasikan, akan langsung dikeluarkan hasil plotnya (1 plot saja di 1 figure). Namun tetap bisa diatasi apabila hendak memplot lebih dari 1 plot pada 1 figure dengan dituliskan lagi data plot lainnya. Misalnya **plt.plot(x1, y1, x2, y2, dst)**.
3. Bisa di dalam ipython notebook --> ipython namun di dalam web (web-based) ipython. Bedanya sudah jelas, kalau ipython notebook itu ada di dalam web (seperti jupyter notebook). Selain itu di ipython notebook juga, tidak hanya bisa menulis code namun juga bisa menggabungkan code dengan berbagai hal misalnya HTML.

Sehingga plt.show() ini akan berguna jika dijalankannya pada media di script dan iptyhon. Di ipython notebook, plt.show() tidak terlalu berguna/tidak ada bedanya. Plot akan tetap ditampilkan jika di ipython notebook.
