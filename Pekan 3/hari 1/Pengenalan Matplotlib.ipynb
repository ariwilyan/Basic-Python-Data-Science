## Pengenalan Matplotlib

**_Matplotlib_** adalah suatu liblary yang **`low-level`** untuk membangun objek visual. maksudnya low-level bukan jelek ya. tapi low-level ini memberikan kita control sampai ke komponen-komponen kecilnya. kalau misalkan kita buat mobil, kita mengontrol sampai pembuatan ke baut bautnya. Oleh karena itu matplotlib penting di fahami sebagai dasar apabila ingin mendalami bidang visualisasi data.

### Komponen Besar : Figure dan Axes

Pada saat ini akan dibahas top-level component, yaitu komponen terbesar dalam suatu objek visualisasi, yaitu **Figure** dan **Axes**.

```python
# contoh sederhana
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = [random.random()*100 for i in range(100)]

# plot
plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.show()
```

Penjelasan untuk komponen terbesar dalam objek visual data.

- **Figure** : window atau page atau halaman dalam objek visual. kalau kita ngegambar di kertas, maka kertas tersebutlah yang di namakan figure.
- **Axis** : kedalam Figure kita menambahkan Axis, Axis adalah suatu area di dalam figure dimana data akan di plot. Dalam Axis juga terdapat berbagai macam komponen lagi seperti x-axis, y-axis, dan lain sebagainya.

```python
# contoh pembuatan visualisasi data sederhana dengan method figure dan axes
import matplotlib.pyplot as plt

fig = plt.figure()
print(fig)
```

Perhatikan kita mempunyai **objek Figure** dengan **0 Axis**. sekarang kita akan menambahkan **axis** kedalam **objek figure** kita dengan method **add_subplot()**. kemudian kita tunjukan objek figure kita dengan method **show()**.

```python
# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menampilkan objek visual
plt.show()
```

Sekarang kita akan tambahkan **data** kedalam **objek tersebut** dengan memanggil method **plot()** dan memberikan **data** sebagai **argument** method tersebut

### Cara 1 (eksplisit)

```python
# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# membuat data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

# menambahkan data ke objek axes
ax.plot(data_x, data_y)

# menampilkan data
plt.show()
```

Seperti itulah cara **sederhana** untuk membuat **objek visualisasi**, sekarang akan di tunjukan **cara lain** yang lebih **clean**.

### Cara 2 (implisit)

```python
# membuat objek visual
plt.plot(data_x, data_y)

# menampilkan objek visual
plt.show()
```

**_Perhatikan_**, **kedua cara** tadi menghasilkan **hasil** yang **sama**. tapi **cara kedua** lebih **clean** karena tidak memerlukan banyak kode. sebenarnya ke dua cara tersebut sama-sama menghasilkan **komponen Figure** dan **Axis**. tetapi **cara pertama Explicit** sedangkan **cara kedua Implicit**.

**Note**:

Cara yang **Explicit** di rekomendasikan ketika kita ingin membuat **visualisasi data** yang **lebih kompleks**, karena kita dapat mengontrol berbagai macam komponen yang terdapat di dalam objek figure dan axes.

Namun, kita juga dapat membuat cara visualisasi data lebih simple dengan menggabungkan pembuatan objek Figure dan Axes dengan sekali jalan, melalui pemanggilan method **subplots()**.

```python
# contoh code dengan penggabungan pembuatan objek figure dan axes dengan method subplots()

## membuat objek figure dan axes
fig, ax = plt.subplots()

## membuat data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

## menambahkan data ke objek axes
ax.plot(data_x, data_y)

## menampilkan data
plt.show()
```

### Membuat plot visualisasi lebih dari satu data ke dalam suatu axis

```python
fig, ax = plt.subplots()

data_x1 = [1, 2, 3, 4]
data_y1 = [10, 20, 25, 30]

data_x2 = [1, 2, 3, 4, 5]
data_y2 = [5, 15, 20, 25, 30]

ax.plot(data_x1, data_y1)
ax.plot(data_x2, data_y2)

plt.show()
```
