<h2>Soal 2: Built-in Function</h2>

Jawab Pertanyaan di bawah ini:
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

//isi jawaban text disini

built-in Function merupakan sebuah fungsi bawaan yang memang sudah tersedia atau disediakan oleh Python untuk membantu pekerjaan kita dalam membuat aplikasi / coding di Python menjadi lebih cepat, praktis dan efisien.

Contoh 3 built-in Function di Python:
- len()
- max()
- min()

```python
# Contoh penggunaan built in function
myList = [6, 8, 10, 5, 7, 3, 4, 1]

# penggunaan built-in Function pertama yaitu len() --> untuk mendapatkan ukuran atau total value yang ada di dalam suatu object
print(f'Total ukuran list: {len(myList)}')
# penggunaan built-in Function kedua yaitu max() --> untuk mendapatkan nilai terbesar
print(f'Nilai maximum yang ada dari list: {max(myList)}')
# penggunaan built-in Function kedua yaitu min() --> untuk mendapatkan nilai terkecil
print(f'Nilai minimum yang ada dari list: {min(myList)}')
```

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

=== Jawaban ===

Method memiliki ciri:
- Dependent (terasosiasi dengan suatu objek) --> sederhananya fungsi dari suatu objek
- Bersifat Implisit
- Dapat beroperasi pada *Python Class*

sedangkan, Function memiliki ciri:
- Independent (tidak terasosiasi dengan suatu objek)
- Bersifat Eksplisit
- Tidak berurusan dengan *Python Class*

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    # isikan code
    return sum(inp_list)/len(inp_list)
    
print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

def kali_list(list1, list2):
    # isikan kode
    return list1 + list2
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
